# pandas的分层索引MultiIndex

分层索引：在一个轴向上拥有多个索引层级，可以表达更高维度数据的形式  
可以更加方便地进行数据筛选，如果有序则性能更好  
groupby等操作结果，如果是多key，结果是分层索引，需要使用   
一般不需要自己船舰分层索引

```python
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv(r'C:\Users\86158\scikit_learn\pollution.csv')
data.head()
```

```python
data['month'] = data['date'].str[:7]
```

```python
data['wnd_dir'].unique()
```

```python
data['snow'].unique()
```

```python
# Series地分层索引MultiIndex
serise_group = data.groupby(['wnd_dir','month'])['temp'].mean()
serise_group.head(10)
```

```python
serise_group.index
```

```python
serise_group.unstack()   # 把二级索引变成列
```

```python
serise_group.reset_index()   # 索引变成列
```

```python
# series有多级索引MultiIndex怎样筛选数据
serise_group.loc['NE'].head()
```

```python
serise_group.loc[('NE','2010-01')]  # 多级索引查询
```

```python
serise_group.loc[:,'2010-01']
```

```python
# DataFrame的多层索引MultiIndex
data.set_index(['wnd_dir','snow'],inplace=True)
data
```

```python
data.sort_index(inplace=True)
data
```

```python
data.index
```

#### DataFrame有多层索引MultiIndex怎样筛选数据
1. 元组(key1,key2)代表筛选多层索引，其中key1是索引第一层，key2代表是第二级
2. 列表[key1,key2]代表同一层的多个key，其中key1，key2是并列的统计索引

```python
data.loc[('cv',14),:]
```

```python
data.loc[['NE','cv'],:]
```

```python
# 两种方式混合,索引值都可以是单个值或者列表
data.loc[(['NE','cv'],0),:]
```

```python
# slice(None)代表筛选这一索引的所有内容
data.loc[(slice(None),[1,2]),:]
```

```python
data.reset_index()  # 多级索引变成普通列
```
