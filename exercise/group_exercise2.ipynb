```python
import pandas as pd
```

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
users.head()
```

```python
# Discover the Male ratio per occupation and sort it from the most to the least
# 使用map将gender字符类型转化成数字类型
gender_dict = {'M':1,'F':0}
users['gender_n'] = users['gender'].map(gender_dict)
male_ratio = users.groupby('occupation')['gender_n'].sum() / users.occupation.value_counts() * 100      # ???
male_ratio.sort_values(ascending=False)
# users.occupation.value_counts()  每个职业的总人数
```

```python
# For each occupation, calculate the minimum and maximum ages
users.groupby('occupation')['age'].agg(['min','max'])
```

```python
# For each combination of occupation and gender, calculate the mean age
users.groupby(['occupation','gender']).age.mean()
```

```python
# For each occupation present the percentage of women and men
gender_count = users.groupby(['occupation','gender']).agg({'gender':'count'})  # DataFrame
occupation_count = users.groupby('occupation').agg('count')
gender_occupation = gender_count.div(occupation_count,level='occupation')  # level:在一个级别上广播，在传递的MultiIndex级别上匹配索引值
gender_occupation['gender']
```

Pandas dataframe.div()类似于datafram/other，但提供了额外的支持来处理输入数据之一中的缺失值。

用法： DataFrame.div (other, axis='columns', level=None, fill_value=None)

参数：
other:DataFrame或常量
axis:对于other输入，轴与other索引匹配
fill_value:用该值填充缺失的(NaN)值。如果两个DataFrame位置都丢失，则结果将丢失
level:在一个级别上广播，在传递的MultiIndex级别上匹配索引值

返回：结果：DataFrame
